# Further Ablation Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
# Read the CSV file
csv_path = './fm_ablation.csv'
df = pd.read_csv(csv_path)


model_names = df.iloc[:, 0].tolist()
x_positions = range(len(df.columns) - 1)

def extract_mean(val):
    return float(val.split('')[0])

mean_df = df.iloc[:, 1:].applymap(extract_mean)

#  Extended colorblind-safe palette 
colorblind_colors = [
    "#1E88E5",  # Blue
    "#FFC107",  # Vermillion
    "#D81B60",  # Magenta
    "#009E73",  # Teal
    "#CC79A7",  # Purple
]
markers = ['o', 's', '^', 'D', 'v', 'P', 'X', '*']

#  Plotting 
plt.figure(figsize=(9, 6.5), dpi=300)
for idx, model in enumerate(model_names):
    y = mean_df.iloc[idx].values
   # if it's the first line, make it bolder:
    if idx == 0:
        lw = 5.0
        z = 10
        effects = [pe.Stroke(linewidth=lw+1, foreground='#1E88E5'), pe.Normal()]
    else:
        lw = 2.5
        z = 5
        effects = None

    line, = plt.plot(
        x_positions, y,
        label=model,
        marker=markers[idx % len(markers)],
        linewidth=lw,
        markersize=6,
        color=colorblind_colors[idx % len(colorblind_colors)],
        zorder=z
    )
    if effects:
        line.set_path_effects(effects)

# Custom x-tick labels
plt.xticks(
    ticks=list(x_positions),
    labels=['0%', '5%', '15%', '30%'],
    fontsize=17
)
plt.xlabel('Missing Percentage', fontsize=17)
plt.ylabel('C-index', fontsize=17)
plt.title('MedImageInsight + Varying Histopathology FMs', fontsize=18, pad=12)

plt.tick_params(axis='y', labelsize=17)

# Overlay legend inside the plot
leg = plt.legend(
    title='Foundation Model',
    title_fontsize=15,
    fontsize=13.5,
    loc='upper right',            # automatically find best spot
    frameon=True,          
    framealpha=0.8,        # slightly transparent
    edgecolor='gray'
)
# bold only the first label
leg.get_texts()[0].set_fontweight('bold')
plt.grid(which='major', linestyle=':', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.savefig("foundation_models.pdf")
plt.show()

```

# Plots for executive summary

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe

# Load data
csv_path = './fm_ablation.csv'
df = pd.read_csv(csv_path)
model_names = df.iloc[:, 0].tolist()
x_positions = range(len(df.columns) - 1)

def extract_mean(val):
    return float(val.split('')[0])
mean_df = df.iloc[:, 1:].applymap(extract_mean)

# Colors & markers
colorblind_colors = ["#1E88E5", "#FFC107", "#D81B60", "#009E73", "#CC79A7"]
markers = ['o', 's', '^', 'D', 'v', 'P', 'X', '*']

#  Adapted Plot 
plt.figure(figsize=(2.5, 2.2), dpi=300)  # Very small figure

for idx, model in enumerate(model_names):
    y = mean_df.iloc[idx].values
    lw = 3.0 if idx == 0 else 1.5
    z = 10 if idx == 0 else 5
    effects = [pe.Stroke(linewidth=lw+1, foreground='#1E88E5'), pe.Normal()] if idx == 0 else None

    line, = plt.plot(
        x_positions, y,
        label=model,
        marker=markers[idx % len(markers)],
        linewidth=lw,
        markersize=3.5,
        color=colorblind_colors[idx % len(colorblind_colors)],
        zorder=z
    )
    if effects:
        line.set_path_effects(effects)

# Axis ticks
plt.xticks(ticks=list(x_positions), labels=['0%', '5%', '15%', '30%'], fontsize=8)
plt.yticks(fontsize=8)

# Legend
leg = plt.legend(
    fontsize=8,
    loc='best',
    #bbox_to_anchor=(0.62, 0.65),
    frameon=True,
    handlelength=1.5,
    borderpad=0.2,
    labelspacing=0.3,
    columnspacing=0.5,
    ncol=1
)
plt.title("MedImageInsight + Varying FMs",fontsize=8)
plt.tight_layout(pad=0.1)
plt.savefig("fms_ablation.pdf")
plt.show()
```
